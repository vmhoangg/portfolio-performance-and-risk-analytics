```python
import pandas as pd
import numpy as np
from pathlib import Path

perf = pd.read_parquet("data/analytics/performance_metrics.parquet")
ret_p = perf["ret_p"]

def var_hist(ret, alpha=0.95, win=252):
    return ret.rolling(win).quantile(1 - alpha)

def es_hist(ret, alpha=0.95, win=252):
    es_vals = []
    for i in range(len(ret)):
        if i < win - 1:
            es_vals.append(np.nan)
            continue
        window = ret.iloc[i - (win - 1): i + 1]
        thr = np.quantile(window, 1 - alpha)
        es_vals.append(window[window <= thr].mean())
    return pd.Series(es_vals, index=ret.index)

var95 = var_hist(ret_p)
es95 = es_hist(ret_p)

risk = pd.DataFrame({
    "VaR95_hist": var95,
    "ES95_hist": es95
})
Path("data/analytics").mkdir(parents=True, exist_ok=True)
risk.to_parquet("data/analytics/risk_metrics.parquet")

risk.tail()
```
