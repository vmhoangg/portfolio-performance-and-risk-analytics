```python
# 04_fetch_sector_benchmark.py
from pathlib import Path
import pandas as pd
import yfinance as yf

SECTOR_ETF = ['XLK','XLY','XLC','XLE','XLF','XLV','XLI','XLP','XLU','XLRE','XLB']
START='2018-01-01'; END='2024-12-31'

raw_dir = Path('data/raw'); raw_dir.mkdir(parents=True, exist_ok=True)
out_prices = raw_dir/'sector_etf_adjclose.parquet'
out_rets   = raw_dir/'sector_etf_returns.parquet'

print("Downloading sector ETFs:", SECTOR_ETF)
df = yf.download(SECTOR_ETF, start=START, end=END, progress=False, auto_adjust=False, group_by='column')
df = df['Adj Close'].copy()
df.index = pd.to_datetime(df.index).tz_localize(None).normalize()
df = df[~df.index.duplicated(keep='last')]
df.to_parquet(out_prices)

rets = df.pct_change()
rets.to_parquet(out_rets)
print("Saved:", out_prices.resolve(), out_rets.resolve())
```
